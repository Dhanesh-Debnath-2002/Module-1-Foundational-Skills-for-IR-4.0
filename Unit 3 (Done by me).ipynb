#### Line Graph

```python
x=[1,2,3]
y=[4,5,2]
plt.plot(x,y) # plotting the points
plt.xlabel('x-label') # naming the x axis
plt.ylabel('y-label') # naming the y axis
plt.title('My first graph') # giving a title to a graph
plt.show() # function to show the plot
```

```python
import numpy as np
x = np.arange(0,10,0.1)
y = np.sin(x)

# figures in matplotlib using figure()
plt.figure(figsize = (5,3))
plt.plot(x,y)
plt.show()
```

#### Subplot

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 7, 1)
plt.plot(x,y)

#plot 2:
x1 = np.array([0, 1, 2, 3])
y1 = np.array([10, 20, 30, 40])

plt.subplot(1, 7, 2)
plt.plot(x1,y1)

#plot 3:
x2 = np.array([0, 1, 2, 3])
y2 = np.array([14, 28, 31, 44])

plt.subplot(1, 7, 3)
plt.plot(x2,y2)

#plot 4:
x3 = np.array([0, 1, 2, 3])
y3 = np.array([14, 8, 1, 44])

plt.subplot(1, 7, 4)
plt.plot(x3,y3)

#plot 5:
x4 = np.array([0, 1, 2, 3])
y4 = np.array([14, 45, 31, 50])

plt.subplot(1, 7, 5)
plt.plot(x4,y4)

#plot 6:
x5 = np.array([0, 1, 2, 3])
y5 = np.array([14, 41, 31, 5])

plt.subplot(1, 7, 6)
plt.plot(x5,y5)

#plot 7:
x6 = np.array([0, 1, 2, 3])
y6 = np.array([10, 45, 31, 7])

plt.subplot(1, 7, 7)
plt.plot(x6,y6)

plt.subplots_adjust(left=0.4,bottom=0.3, right=2, top=2 ,wspace=0.4, hspace=0.4) #provide space between subplots
plt.show()
```

#### Annotation

```python
fig, ax= plt.subplots()
x = np.linspace(0, 20, 1000)
ax.plot(x, np.cos(x))
ax.axis("equal")
#text annotation to the plot where it indicate maximum value of the curve
ax.annotate('local maximum' , xy=(6.28,1),xytext=(10,4), arrowprops=dict(facecolor = 'black', shrink=0.05))
#text annotation to the plot where it indicate minimum value of the curve
ax.annotate('local minimum' , xy=(5 * np.pi ,-1),xytext = (2,-6),arrowprops = dict(arrowstyle="->",connectionstyle= "angle3,angleA=0,angleB=-90"))
plt.show()
```

```python
fig, ax= plt.subplots()
x = np.linspace(0, 20, 500)
ax.plot(x, np.cos(x))
ax.axis("equal")
#text annotation to the plot where it indicate maximum value of the curve
ax.annotate('local maximum' , xy=(12,1),xytext=(10,4), arrowprops=dict(facecolor = 'black', shrink=4.25))
#text annotation to the plot where it indicate minimum value of the curve
ax.annotate('local minimum' , xy=(5*(22/7),-1),xytext = (50,-6),arrowprops = dict(arrowstyle="->",connectionstyle= "angle3,angleA=0,angleB=-90"))
plt.show()
```

#### DPI

```python
w=4
h=3
d=200
plt.figure(figsize=(w,h),dpi =d)
plt.axis([0,5,0,5])
x=[0,3,5]
y=[1,4,3.5]
label_x = 1
label_y = 4
arrow_x = 3
arrow_y = 4
arrow_properties = dict(facecolor="black",width =0.5,headwidth=4, shrink=0.1)
plt.annotate("maximum",xy=(arrow_x,arrow_y), xytext=(label_x,label_y),arrowprops=arrow_properties)
plt.plot(x,y)
plt.show()
```

#### 3D plot

```python
#importing mplot3d tootlkits
from mpl_toolkits import mplot3d
```

```python
fig = plt.figure()

#syntax for 3d projection
ax = plt.axes(projection='3d')

#defining all 3 axes
z=np.linspace(0,1,100)
x=z*np.sin(25*z)
y=z*np.cos(25*z)

#plotting
ax.plot3D(x,y,z,'green')
ax.set_title('3D line plot')
plt.show()
```
